Input : (’12’, ’21’)
Output : ’110’

````python
def add_ternary(a, b):
    # Convert from base 3 to integer
    num1 = int(a, 3)
    num2 = int(b, 3)

    # Sum the two integers
    total = num1 + num2

    return np.base_repr(total, base=3)


# Example usage:
input1, input2 = '12', '21'
result = add_ternary(input1, input2)
print(result)
````

6. You are given two non-empty linked lists representing two non-negative integers. The digits are stored in
reverse order and each of their nodes contain a single digit. Add the two numbers and return it as a linked
list. You may assume the two numbers do not contain any leading zero, except the number 0 itself.
1
e.g.,
Input: (2 -> 4 -> 3) + (5 -> 6 -> 4)
Output: 7 -> 0 -> 8
Explanation: 342 + 465 = 807.
Linked list is defined as follow
```python
# Definition for singly-linked list.
# class ListNode:
# def __init__(self, x):
# self.val = x
# self.next = None
```

````python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None


def addTwoNumbers(l1: ListNode, l2: ListNode) -> ListNode:
    dummy = ListNode(0)
    current = dummy
    carry = 0

    # Loop until both lists are exhausted and no carry remains.
    while l1 or l2 or carry:
        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0

        total = v1 + v2 + carry
        carry = total // 10  # update carry for next iteration
        current.next = ListNode(total % 10)
        current = current.next

        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    return dummy.next

# Helper function to print linked list.


def printLinkedList(node: ListNode):
    while node:
        print(node.val, end="")
        if node.next:
            print(" -> ", end="")
        node = node.next
    print()  # for a newline

# Test code to verify the algorithm.


l1 = ListNode(2)
l1.next = ListNode(4)
l1.next.next = ListNode(3)
l2 = ListNode(5)
l2.next = ListNode(6)
l2.next.next = ListNode(4)
result = addTwoNumbers(l1, l2)
print("Output:", end=" ")
printLinkedList(result)
````

10. Implement heap sort

````python
import copy
import heapq


def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap if the element found is greater than the next element
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):
    merged = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # Append remaining elements
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged


def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


def heap_sort(arr):
    heap = []
    for element in arr:
        heapq.heappush(heap, element)
    sorted_arr = []
    while heap:
        sorted_arr.append(heapq.heappop(heap))
    return sorted_arr


test_data = [64, 34, 25, 12, 22, 11, 90]
print("Original data:", test_data)
# Bubble sort (in-place sort)
bubble_data = copy.deepcopy(test_data)
sorted_bubble = bubble_sort(bubble_data)
print("Bubble sort result: ", sorted_bubble)
# Merge sort (returns a new sorted list)
sorted_merge = merge_sort(copy.deepcopy(test_data))
print("Merge sort result:  ", sorted_merge)
# Quick sort (returns a new sorted list)
sorted_quick = quick_sort(copy.deepcopy(test_data))
print("Quick sort result:  ", sorted_quick)
# Heap sort (returns a new sorted list)
sorted_heap = heap_sort(copy.deepcopy(test_data))
print("Heap sort result:   ", sorted_heap)
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt


def linear_regression():
    # Create synthetic data for y = 2x + 3 with noise
    torch.manual_seed(42)
    X = torch.unsqueeze(torch.linspace(-10, 10, 100), dim=1)  # shape: [100, 1]
    y = 2 * X + 3 + torch.randn(X.size()) * 2  # add noise

    # Define a simple linear regression model
    model = nn.Linear(1, 1)
    criterion = nn.MSELoss()  # Mean Squared Error Loss
    optimizer = optim.SGD(model.parameters(), lr=0.01)

    # Training loop
    epochs = 2000
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()        # Zero the gradients
        outputs = model(X)           # Forward pass
        loss = criterion(outputs, y)  # Compute loss
        loss.backward()              # Backward pass (autograd computes gradients)
        optimizer.step()             # Update parameters

        # Print progress every 500 epochs
        if (epoch + 1) % 500 == 0:
            print(
                f'Linear Regression Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

    # Print learned parameters
    learned_weight = list(model.parameters())[0].item()
    learned_bias = list(model.parameters())[1].item()
    print("Learned parameters for Linear Regression:")
    print("  Weight:", learned_weight)
    print("  Bias:", learned_bias)

    # plot
    plt.scatter(X, y)
    plt.plot(X, learned_weight*X + learned_bias, color='red')
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title("Linear Regression")
    plt.show()

def logistic_regression():
    # Create synthetic binary classification data
    torch.manual_seed(42)
    N = 100
    X = torch.randn(N, 2)
    # Generate labels: label = 1 if sum of features > 0, else 0
    y = (X.sum(dim=1) > 0).float().unsqueeze(1)

    # Define logistic regression model
    model = nn.Linear(2, 1)
    # Use BCEWithLogitsLoss (it applies sigmoid internally)
    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.SGD(model.parameters(), lr=0.1)

    # Training loop
    epochs = 2000
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()       # Zero the gradients
        logits = model(X)           # Forward pass (raw scores)
        loss = criterion(logits, y)  # Compute binary cross entropy loss
        loss.backward()             # Backward pass
        optimizer.step()            # Update parameters

        # Print progress every 500 epochs
        if (epoch + 1) % 500 == 0:
            print(
                f'Logistic Regression Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

    # Evaluate model accuracy
    with torch.no_grad():
        logits = model(X)
        predictions = torch.sigmoid(logits) > 0.5
        accuracy = (predictions.float() == y).float().mean()
        print("Logistic Regression accuracy:", accuracy.item())
        
    # plot
    plt.scatter(X[:, 0], X[:, 1], c=y.squeeze())
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.title("Logistic Regression")
    plt.show()


print("=== Linear Regression Example ===")
linear_regression()
print("\n=== Logistic Regression Example ===")
logistic_regression()
````

Implement linear SVM model for binary classification task and use autograd to optimize it by Pytorch.
Hint: you may use the loss of $\sum max(0, 1- y(wx+b))$

````python
import torch
import torch.nn as nn
import torch.optim as optim


def hinge_loss(outputs, labels):
    """
    Compute the average hinge loss.
    outputs: [N, 1], the model's raw output (w^T x + b)
    labels:  [N, 1], values in {-1, +1}
    """
    # Hinge loss = mean( max(0, 1 - labels * outputs) )
    return torch.mean(torch.clamp(1 - labels * outputs, min=0))


def main():
    torch.manual_seed(42)

    # --------------------------
    # 1. Create Synthetic Data
    # --------------------------
    # We'll create a 2D dataset of N points.
    # Label = +1 if x1 + x2 > 0, else -1.
    N = 200
    X = torch.randn(N, 2)  # shape [N, 2]
    y = torch.sign(X.sum(dim=1)).view(-1, 1)  # shape [N, 1], in {-1, 0, +1}
    # If sum is exactly zero, sign gives 0. Let's force 0 to +1 for simplicity.
    y[y.eq(0)] = 1

    # --------------------------
    # 2. Define the Linear Model
    # --------------------------
    # Our SVM is simply a linear layer that outputs w^T x + b.
    model = nn.Linear(2, 1, bias=True)

    # --------------------------
    # 3. Set Up Optimizer
    # --------------------------
    # We'll use simple SGD. Feel free to experiment with other optimizers.
    optimizer = optim.SGD(model.parameters(), lr=0.01)

    # --------------------------
    # 4. Training Loop
    # --------------------------
    epochs = 1000
    for epoch in range(epochs):
        # Zero out gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(X)  # shape [N, 1]

        # Compute hinge loss
        loss = hinge_loss(outputs, y)

        # Backward pass (autograd)
        loss.backward()

        # Update parameters
        optimizer.step()

        # Print progress every 200 epochs
        if (epoch + 1) % 200 == 0:
            print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

    # --------------------------
    # 5. Evaluate Model
    # --------------------------
    with torch.no_grad():
        # Get raw outputs
        outputs = model(X)
        # Predict label by sign of output: +1 if > 0, -1 if <= 0
        preds = torch.sign(outputs)
        # Replace any zero sign with +1
        preds[preds.eq(0)] = 1

        # Compute accuracy
        correct = (preds == y).sum().item()
        accuracy = correct / N
        print(f"Training Accuracy: {accuracy * 100:.2f}%")

        # plot
        plt.scatter(X[:, 0], X[:, 1], c=y.squeeze())
        plt.xlabel("X1")
        plt.ylabel("X2")
        plt.title("SVM")
        plt.show()


main()
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt


def hinge_loss(outputs, labels):
    """
    Compute the average hinge loss.
    outputs: [N, 1], the model's raw output (w^T x + b)
    labels:  [N, 1], values in {-1, +1}
    """
    # Hinge loss = mean( max(0, 1 - labels * outputs) )
    return torch.mean(torch.clamp(1 - labels * outputs, min=0))


def main():
    torch.manual_seed(42)

    # --------------------------
    # 1. Create Synthetic Data
    # --------------------------
    # We'll create a 2D dataset of N points.
    # Label = +1 if x1 + x2 > 0, else -1.
    N = 200
    X = torch.randn(N, 2)  # shape [N, 2]
    y = torch.sign(X.sum(dim=1)).view(-1, 1)  # shape [N, 1], in {-1, 0, +1}
    # If sum is exactly zero, sign gives 0. Let's force 0 to +1 for simplicity.
    y[y.eq(0)] = 1

    # --------------------------
    # 2. Define the Linear Model
    # --------------------------
    # Our SVM is simply a linear layer that outputs w^T x + b.
    model = nn.Linear(2, 1, bias=True)

    # --------------------------
    # 3. Set Up Optimizer
    # --------------------------
    # We'll use simple SGD. Feel free to experiment with other optimizers.
    optimizer = optim.SGD(model.parameters(), lr=0.01)

    # Regularization strength
    reg_lambda = 0.01

    # --------------------------
    # 4. Training Loop
    # --------------------------
    epochs = 1000
    for epoch in range(epochs):
        optimizer.zero_grad()

        # Forward pass
        outputs = model(X)  # shape [N, 1]

        # -- (a) Compute hinge loss
        loss = hinge_loss(outputs, y)

        # -- (b) Add Frobenius norm penalty in two different ways:

        # 1) Using PyTorch's built-in norm function (p='fro' gives Frobenius norm)
        # fro_norm_1 = torch.norm(model.weight, p='fro')
        # loss += reg_lambda * fro_norm_1

        # 2) Implementing the Frobenius norm yourself:
        fro_norm_2 = torch.sqrt(torch.sum(model.weight ** 2))
        loss += reg_lambda * fro_norm_2

        # Backward pass
        loss.backward()
        optimizer.step()

        # Print progress every 200 epochs
        if (epoch + 1) % 200 == 0:
            print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

    # --------------------------
    # 5. Evaluate Model
    # --------------------------
    with torch.no_grad():
        outputs = model(X)
        preds = torch.sign(outputs)
        # Replace any zero sign with +1
        preds[preds.eq(0)] = 1

        correct = (preds == y).sum().item()
        accuracy = correct / N
        print(f"Training Accuracy: {accuracy * 100:.2f}%")

        # Plot the data
        plt.scatter(X[:, 0], X[:, 1], c=y.squeeze(), cmap='bwr')
        plt.xlabel("X1")
        plt.ylabel("X2")
        plt.title("SVM with Frobenius Norm Penalty")
        plt.show()


main()
````

````python
import numpy as np

# 1. Convert Cartesian to Polar Coordinates
# Generate a random 10x2 matrix (each row: [x, y])
cartesian = np.random.rand(10, 2)
# Compute r and theta
r = np.sqrt(np.sum(cartesian**2, axis=1))
theta = np.arctan2(cartesian[:, 1], cartesian[:, 0])
polar = np.column_stack((r, theta))
print("Cartesian Coordinates (10x2):\n", cartesian)
print("\nPolar Coordinates (r, theta):\n", polar)

# 2. Create a 2D Array Subclass such that Z[i, j] == Z[j, i]


class SymmetricArray(np.ndarray):
    def __new__(cls, input_array):
        obj = np.asarray(input_array).view(cls)
        if obj.ndim != 2:
            raise ValueError("SymmetricArray must be 2D")
        # Enforce symmetry: symmetrize the array
        obj = (obj + obj.T) / 2
        return obj

    def __array_finalize__(self, obj):
        if obj is None:
            return


# Test the SymmetricArray subclass
Z = SymmetricArray(np.random.rand(5, 5))
print("\nSymmetric Array Z (5x5):\n", Z)
print("\nDifference Z - Z.T (should be near zero):\n", Z - Z.T)

# 3. Compute Distance from Each Point to Each Line
# Define random sets of points for lines and query points
num_lines = 3
num_points = 4
P0 = np.random.rand(num_lines, 2)  # Starting point of each line
P1 = np.random.rand(num_lines, 2)  # Ending point of each line
P = np.random.rand(num_points, 2)  # Query points


def point_to_line_distance(P0, P1, P):
    """
    Compute distances from each point in P to each line defined by (P0[i], P1[i]).
    Returns a distance matrix of shape (num_lines, num_points).
    """
    # Vector along each line
    L = P1 - P0              # shape: (num_lines, 2)
    norm_L = np.linalg.norm(L, axis=1)  # shape: (num_lines,)

    # Compute vector from P0 to every point in P for each line
    # Broadcasting to get shape (num_lines, num_points, 2)
    V = P[None, :, :] - P0[:, None, :]

    # Expand L to shape (num_lines, 1, 2) for broadcasting
    L_exp = L[:, None, :]

    # Compute cross product magnitude in 2D: |L x V| = |L_x * V_y - L_y * V_x|
    cross = np.abs(L_exp[..., 0] * V[..., 1] - L_exp[..., 1] * V[..., 0])

    # Compute distance = cross / norm(L)
    distances = cross / norm_L[:, None]
    return distances


distances = point_to_line_distance(P0, P1, P)
print("\nLine Endpoints P0:\n", P0)
print("\nLine Endpoints P1:\n", P1)
print("\nQuery Points P:\n", P)
print("\nDistances from each point to each line (rows: lines, columns: points):\n", distances)
````

## 23. Bilinear Interpolation

Implement the bilinear interpolation algorithm in Python.

### Problem Description

Given a 2D array representing image pixel intensities, implement a function `BilinearInterpolation(A, p)` that computes the interpolated intensity at point `p = (x, y)`. Assume that the input coordinates are 1-indexed, meaning that the pixel at position (1, 1) corresponds to the top-left element of `A`.

**Test Samples:**

For the matrix

$$
A =
\begin{pmatrix}
110 & 120 & 130 \\
210 & 220 & 230 \\
310 & 320 & 330
\end{pmatrix}
$$

- `BilinearInterpolation(A, (1, 1))` should return **110**.
- `BilinearInterpolation(A, (2.5, 2.5))` should return **275**.

### Method

1. Convert the input coordinates from 1-indexed to 0-indexed.
2. Identify the four neighboring pixels:
   - Let $(x_1, y_1)$ be the floor of the coordinates.
   - Let $(x_2, y_2)$ be the ceiling of the coordinates.
3. Compute the fractional parts $dx$ and $dy$.
4. Apply the bilinear interpolation formula:
   $$
   f(x, y) = Q_{11} (1-dx)(1-dy) + Q_{21} \, dx (1-dy) + Q_{12} (1-dx) \, dy + Q_{22} \, dx \, dy
   $$
   where $Q_{11}$, $Q_{21}$, $Q_{12}$, and $Q_{22}$ are the values at the four neighboring pixels.
5. Return the interpolated value.


````python
import numpy as np


def BilinearInterpolation(A, point):
    """
    Perform bilinear interpolation on a 2D array A at the given point.

    Parameters:
        A (list of lists or 2D numpy array): Input matrix with pixel intensities.
        point (tuple): A tuple (x, y) representing the point (1-indexed).

    Returns:
        float: Interpolated value at the given point.
    """
    # Convert A to a numpy array if not already
    A = np.array(A, dtype=float)

    # Unpack the 1-indexed point and convert to 0-indexed coordinates
    x, y = point
    x, y = x - 1, y - 1  # now 0-indexed

    # Get the integer parts (floors) and ceilings
    x1, y1 = int(np.floor(x)), int(np.floor(y))
    x2, y2 = int(np.ceil(x)), int(np.ceil(y))

    # If the point is exactly on a pixel, return its value directly.
    if x1 == x2 and y1 == y2:
        return A[y1, x1]

    # Compute the fractional parts
    dx, dy = x - x1, y - y1

    # Get the four neighboring pixel values
    Q11 = A[y1, x1]
    Q21 = A[y1, x2]
    Q12 = A[y2, x1]
    Q22 = A[y2, x2]

    # Compute the interpolated value using the bilinear interpolation formula
    value = (Q11 * (1 - dx) * (1 - dy) +
             Q21 * dx * (1 - dy) +
             Q12 * (1 - dx) * dy +
             Q22 * dx * dy)
    return value


# Test samples:
A = [
    [110, 120, 130],
    [210, 220, 230],
    [310, 320, 330]
]

# Test 1: BilinearInterpolation(A, (1, 1)) should return 110
result1 = BilinearInterpolation(A, (1, 1))
print("BilinearInterpolation(A, (1, 1)) =", result1)  # Expected output: 110

# Test 2: BilinearInterpolation(A, (2.5, 2.5)) should return 275
result2 = BilinearInterpolation(A, (2.5, 2.5))
print("BilinearInterpolation(A, (2.5, 2.5)) =", result2)  # Expected output: 275
````

## 25. Extracting a Subpart of an Array

Write a function that, given an arbitrary 2D array, extracts a subarray of a fixed shape centered on a given element. If the desired subarray extends beyond the bounds of the original array, pad the missing elements with a specified fill value.

**Example:**

```python
Z = np.array([
    [3, 6, 8, 5, 9],
    [4, 9, 0, 0, 9],
    [6, 1, 4, 0, 8],
    [9, 1, 2, 0, 9],
    [4, 1, 7, 5, 0]
])
shape = (4, 4)
fill = 0
position = (1, 1)

[[0, 0, 0, 0],
 [0, 3, 6, 8],
 [0, 4, 9, 0],
 [0, 6, 1, 4]]
```

````python
import numpy as np


def extract_subarray(Z, shape, position, fill=0):
    """
    Extract a subarray of fixed shape from Z centered on position.
    Pads with the fill value if the window extends beyond Z's boundaries.

    Parameters:
        Z (2D array): Input array.
        shape (tuple): Desired subarray shape (height, width).
        position (tuple): Center position (row, col) in the original array (0-indexed).
        fill: Value used for padding.

    Returns:
        subarray (2D array): Extracted subarray of given shape.
    """
    H, W = shape
    center_row, center_col = position

    # Calculate starting indices in the original array.
    start_row = center_row - H // 2
    start_col = center_col - W // 2

    # Create an output array filled with the fill value.
    subarray = np.full((H, W), fill, dtype=Z.dtype)

    for i in range(H):
        for j in range(W):
            orig_row = start_row + i
            orig_col = start_col + j
            # Check if the index is within the bounds of Z.
            if 0 <= orig_row < Z.shape[0] and 0 <= orig_col < Z.shape[1]:
                subarray[i, j] = Z[orig_row, orig_col]
    return subarray


# Test sample:
Z = np.array([
    [3, 6, 8, 5, 9],
    [4, 9, 0, 0, 9],
    [6, 1, 4, 0, 8],
    [9, 1, 2, 0, 9],
    [4, 1, 7, 5, 0]
])
shape = (4, 4)
fill = 0
position = (1, 1)  # 0-indexed center position

result = extract_subarray(Z, shape, position, fill)
print(result)
````

## 26. Matrix Operations

Implement the following 2D matrix operations **without NumPy**:

- **add(a, b):** Element-wise addition of matrices `a` and `b`.
- **subtract(a, b):** Element-wise subtraction of matrix `b` from `a`.
- **scalar_multiply(matrix, scalar):** Multiply each element of the matrix by a scalar.
- **multiply(a, b):** Standard matrix multiplication.
- **identity(n):** Create an `n×n` identity matrix.
- **transpose(matrix):** Transpose of a matrix.
- **inverse(matrix):** Compute the inverse of a square matrix using Gauss-Jordan elimination.

**Test Samples:**

```python
matrix_a = [[12, 10], [3, 9]]
matrix_b = [[3, 4], [7, 4]]
matrix_c = [[11, 12, 13, 14],
            [21, 22, 23, 24],
            [31, 32, 33, 34],
            [41, 42, 43, 44]]
matrix_d = [[3, 0, 2],
            [2, 0, -2],
            [0, 1, 1]]

add(matrix_a, matrix_b)       == [[15, 14], [10, 13]]
subtract(matrix_a, matrix_b)  == [[9, 6], [-4, 5]]
scalar_multiply(matrix_b, 3)  == [[9, 12], [21, 12]]
multiply(matrix_a, matrix_b)  == [[106, 88], [72, 48]]
identity(3)                 == [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
transpose(matrix_c)         == [[11, 21, 31, 41],
                                [12, 22, 32, 42],
                                [13, 23, 33, 43],
                                [14, 24, 34, 44]]
inverse(matrix_d)           == [[0.2, 0.2, 0.0],
                                [-0.2, 0.3, 1.0],
                                [0.2, -0.3, 0.0]]
```

````python
def add(a, b):
    """Element-wise addition of two matrices."""
    rows = len(a)
    cols = len(a[0])
    return [[a[i][j] + b[i][j] for j in range(cols)] for i in range(rows)]


def subtract(a, b):
    """Element-wise subtraction of matrix b from matrix a."""
    rows = len(a)
    cols = len(a[0])
    return [[a[i][j] - b[i][j] for j in range(cols)] for i in range(rows)]


def scalar_multiply(matrix, scalar):
    """Multiply each element of the matrix by scalar."""
    return [[scalar * element for element in row] for row in matrix]


def multiply(a, b):
    """Standard matrix multiplication: a * b."""
    rows_a = len(a)
    cols_a = len(a[0])
    rows_b = len(b)
    cols_b = len(b[0])
    if cols_a != rows_b:
        raise ValueError("Incompatible dimensions for matrix multiplication")
    # Initialize result matrix with zeros
    result = [[0 for _ in range(cols_b)] for _ in range(rows_a)]
    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result[i][j] += a[i][k] * b[k][j]
    return result


def identity(n):
    """Return an n x n identity matrix."""
    return [[1 if i == j else 0 for j in range(n)] for i in range(n)]


def transpose(matrix):
    """Transpose of a matrix."""
    rows = len(matrix)
    cols = len(matrix[0])
    return [[matrix[i][j] for i in range(rows)] for j in range(cols)]


def inverse(matrix):
    """Compute the inverse of a square matrix using Gauss-Jordan elimination."""
    n = len(matrix)
    # Create an augmented matrix with the identity matrix on the right
    aug = [row[:] + identity(n)[i] for i, row in enumerate(matrix)]

    # Perform Gauss-Jordan elimination
    for i in range(n):
        # Find the pivot (assume non-zero pivot, else error)
        pivot = aug[i][i]
        if pivot == 0:
            raise ValueError("Matrix is singular and cannot be inverted.")
        # Normalize the pivot row
        for j in range(2*n):
            aug[i][j] /= pivot
        # Eliminate the i-th element in all other rows
        for k in range(n):
            if k != i:
                factor = aug[k][i]
                for j in range(2*n):
                    aug[k][j] -= factor * aug[i][j]

    # Extract the inverse matrix from the augmented matrix
    inv_matrix = [row[n:] for row in aug]
    return inv_matrix


# Test samples:
matrix_a = [[12, 10], [3, 9]]
matrix_b = [[3, 4], [7, 4]]
matrix_c = [[11, 12, 13, 14],
            [21, 22, 23, 24],
            [31, 32, 33, 34],
            [41, 42, 43, 44]]
matrix_d = [[3, 0, 2],
            [2, 0, -2],
            [0, 1, 1]]

print("add(matrix_a, matrix_b) =", add(matrix_a, matrix_b))
print("subtract(matrix_a, matrix_b) =", subtract(matrix_a, matrix_b))
print("scalar_multiply(matrix_b, 3) =", scalar_multiply(matrix_b, 3))
print("multiply(matrix_a, matrix_b) =", multiply(matrix_a, matrix_b))
print("identity(3) =", identity(3))
print("transpose(matrix_c) =", transpose(matrix_c))
print("inverse(matrix_d) =", inverse(matrix_d))
````

## 27. Greatest Common Divisor (GCD)

Implement a function to find the greatest common divisor (gcd) of two integers using Euclid's algorithm.

### Test Samples:
- GCD(3, 5) = 1
- GCD(6, 3) = 3
- GCD(-2, 6) = 2
- GCD(0, 3) = 3


````python
def gcd(a, b):
    """
    Compute the greatest common divisor of a and b using Euclid's algorithm.

    Parameters:
        a (int): First integer.
        b (int): Second integer.

    Returns:
        int: The greatest common divisor of a and b.
    """
    # Make sure both numbers are non-negative
    a, b = abs(a), abs(b)

    # Euclid's algorithm
    while b:
        a, b = b, a % b
    return a


# Test samples:
print("GCD(3, 5) =", gcd(3, 5))      # Expected: 1
print("GCD(6, 3) =", gcd(6, 3))      # Expected: 3
print("GCD(-2, 6) =", gcd(-2, 6))    # Expected: 2
print("GCD(0, 3) =", gcd(0, 3))      # Expected: 3
````

## 28. Find All Consecutive Positive Number Sequences Whose Sum is N

Given a positive integer \( N \), find all sequences of consecutive positive integers whose sum equals \( N \).

**Example:**
For \( N = 100 \), possible sequences include:
- \( 18 + 19 + 20 + 21 + 22 = 100 \)
- \( 9 + 10 + 11 + 12 + 13 + 14 + 15 + 16 = 100 \)

**Task:**
Find all consecutive positive number sequences whose sum is \( 1000 \), and report your results.


````python
def find_consecutive_sequences(N):
    """
    Find all sequences of consecutive positive integers that sum to N.

    Uses the formula:
      sum of L consecutive numbers starting from a = L * a + L*(L-1)/2 = N,
      which gives a = (N - L*(L-1)/2) / L.

    Returns:
      A list of sequences, where each sequence is represented as a list of integers.
    """
    sequences = []
    L = 2  # start with sequence lengths of at least 2
    while L * (L + 1) // 2 <= N:
        numerator = N - (L * (L - 1) // 2)
        if numerator % L == 0:  # a must be an integer
            a = numerator // L
            if a > 0:
                seq = list(range(a, a + L))
                sequences.append(seq)
        L += 1
    return sequences


# Find and print all sequences for N = 1000
N = 1000
sequences = find_consecutive_sequences(N)
print(f"\nConsecutive positive number sequences whose sum is {N}:")
for seq in sequences:
    print(seq)
````
